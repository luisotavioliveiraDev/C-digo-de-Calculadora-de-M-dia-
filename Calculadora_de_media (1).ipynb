```
nome = input("digite o nome completo do aluno:")
nota1 = float(input("digite a primeira nota:"))
nota2 = float(input("digite a segunda nota:"))
media = (nota1+nota2)/2

print("media:",media)
if(media>=7): # caso a media seja maior igual sete
  print(nome + " foi aprovado")
else:
  if(media>=3): # mais um if dentro do else dizendo se a media por maior igual a 3
     print(nome +  " está na prova final")
  else: # qualquer outro resultado reprovado
    print(nome +  " foi reprovado")
```
